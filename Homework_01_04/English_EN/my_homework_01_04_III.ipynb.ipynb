From the CSV hospitales2.csv, generate an output CSV file, containing the following fields in this order:
- latitude
- length
-Name
-label<br>
With the corresponding data extracted from the original CSV, the name field has to correspond to the hospital address, and the label field to the hospital name.

```python
import pandas as pd
```

Dataframe creation

```python
hospitals = pd.read_csv("hospitales2.csv",sep=",",encoding="utf-8")
hospitals.head()
```

Cleaning the "WKT" column 

```python

hospitals.WKT = hospitals.WKT.str.replace("POINT","")
hospitals.WKT = hospitals.WKT.str.replace("(","")
hospitals.WKT = hospitals.WKT.str.replace(")","")
hospitals.head()
    
```

Rename column "WKT" to "Coordinates"

```python
hospitals.rename(columns={"WKT": "Coordinates"},inplace=True)
```

Creation of new columns called "Latitude" and "Longitude" to be able to divide and store the data of the Coordinates column<br><br>
Note that it also creates an empty name column since when separating the Coordinates column, it will return three divisions because the space counts as a character.

```python
hospitals[[' ','Latitude', 'Longitude']] = hospitals['Coordinates'].str.split(' ', expand=True)
del hospitals["Coordinates"]
hospitals.head(3)
```

Assignment to the new columns of "Name" and "Label" the values of the columns "DOM_NORMA" and "NAME"

```python
hospitals["Name"] = hospitals["DOM_NORMA"]
hospitals["Label"] = hospitals["NOMBRE"]
```

```python
hospitals.head(1)
```

Finally, the creation of the new Dataframe with the columns "Latitude","Longitude","Name" and "Label" with the order required in the order

```python
new_hospitals = hospitals[["Latitude","Longitude","Name","Label"]]
```
