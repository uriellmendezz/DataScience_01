1) Implementar un juego, que consista en apilar números enteros del 1 al 20, de forma aleatoria, para lo cual debe usarse una estructura de Pila. 
Luego, el usuario debe elegir un número de veces en que se va a quitar elementos de la pila, los cuales, sumados entre sí, no deben superar el valor de 50.
El usuario pierde si la suma supera ese valor. Si no lo supera, gana, pero su calificación será 10 menos el número elementos que falten quitar para todavía no superar 50.
El programa debe informar si perdió, y si ganó, con qué calificación lo hizo.

Consideraciones:<br>
a. Se puede usar la función input() para obtener una entrada de teclado.<br>
b. Se puede usar la el modulo random para obtener valores aleatorios.

```python
import random
import numpy as np
```

```python

class Juego_Pila():
    def __init__ (self):
        self.lista = list(range(1,21))

    def jugar(self):
        random.shuffle(self.lista) #Desordeno la lista aleatoriamente
        user_choice = input("¿Cuantos numeros queres eliminar de la pila?")

        try:
            user_choice = int(user_choice)
            lista_modificada = self.lista.copy() # Creo una copia de la lista para poder eliminarla y no alterar la original
        
            # Elimino la cantidad de numeros que el usuario elija y los almaceno en varibale "numeros_eliminados"
            numeros_eliminados = []
            for i in range(user_choice):
                eliminado = lista_modificada.pop(0)
                numeros_eliminados.append(eliminado)
                    
            # Agrego un espacio vacio en los lugares donde se eliminaron numeros
            for i in range(user_choice):
                lista_modificada.insert(0,"")
                
            lista_completa = self.lista.copy()
            suma = sum(numeros_eliminados)
            
            print("")
            print("Esta era la pila completa")
            print(lista_completa)
            print("")
            print("Lista sin los eliminados")
            print(lista_modificada)
            print("")
            print(numeros_eliminados, "=" ,suma)
            print("---------------------------------------------")
                
            if suma < 50:
                contador = 0
                while suma <= 50:
                    suma += lista_modificada[user_choice + 1]
                    contador += 1
                total = 10 - contador

                if total < 0:
                    total = 0

                print(f"Obtuviste una calificacion de {total}/10")
                    
            elif suma == 50:
                print("Obtuviste el mejor resultado posible. Bien hecho. 10/10")

            else:
                print("Perdiste. Volve a intentarlo...")

        except:
            print("Por favor, coloca un numero")
            self.jugar()

        
```

```python
usuario = Juego_Pila()
usuario.jugar()
```

2) Implementar un juego donde constas de 2 jarras, de capacidad 5 y 3 litros respectivamente, y debes colocar 4 litros en la jarra de 5L.
Las opciones posibles son:
* Llenar la jarra de 3 litros
* Llenar la jarra de 5 litros
* Vaciar la jarra de 3 litros
* Vaciar la jarra de 5 litros
* Verter el contenido de la jarra de 3 litros en la de 5 litros
* Verter el contenido de la jarra de 5 litros en la de 3 litros

```python
class Juego_Jarras():
    def __init__ (self):
        self.jarra3l = 0
        self.jarra5l = 0

    def objetivo(self):
        print("El objetivo del juego es colocar 4 litros en la jarra de 5 litros.")
        
    def acciones(self):
        print("Las acciones disponibles son las siguientes:")
        print("===> Llenar la jarra de 3 litros")
        print("===> Llenar la jarra de 5 litros")
        print("===> Vaciar la jarra de 3 litros")
        print("===> Vaciar la jarra de 5 litros")
        print("===> Verter el contenido de la jarra de 3 litros en la de 5 litros")
        print("===> Verter el contenido de la jarra de 5 litros en la de 3 litros")

    def ayuda(self):
        print("objetivo()  ===> para ver el objetivo del juego")
        print("acciones()  ===> para ver las acciones posibles")
        print("llenar()    ===> para llenar una jarra")
        print("vaciar()    ===> para vaciar una jarra")
        print("verter()    ===> para verter agua en otra jarra")
        print("consultar() ===> para ver el estado de las jarras")

    def llenar(self,jarra):
        if jarra == 3:
            if self.jarra3l > 0:
                print(">>> La jarra de 3 litros ya esta llena")
                self.consultar()
            elif self.jarra3l == 0:
                self.jarra3l += 3
                print(">>> llenar jarra de 3 litros √")
                self.consultar()

        elif jarra == 5:
            if self.jarra5l > 0:
                print(">>> La jarra de 5 litros ya esta llena")
                self.consultar()
            elif self.jarra5l == 0:
                self.jarra5l += 5
                print(">>> Llenar jarra de 5 litros √")
                self.consultar()
        else:
            print(">>> 3 ===> para llenar la jarra de 3 litros\n>>> 5 ===> para llenar la jarra de 5 litros")

    def vaciar(self,jarra):
        if jarra == 3:
            if self.jarra3l == 0:
                print(">>> La jarra de 3 litros ya estaba vacia")
            else:
                self.jarra3l = 0
                print(">>> Vaciar la jarra de 3 litros √")
                self.consultar()
        elif jarra == 5:
            if self.jarra5l == 0:
                print(">>> La jarra de 5 litros ya estaba vacia")
            else:
                self.jarra5l = 0
                print(">>> Vaciar la jarra de 5 litros √")
                self.consultar()
        else:
            print(">>> 3 ===> para vaciar la jarra de 3 litros\n>>> 5 ===> para vaciar la jarra de 5 litros")

    def verter(self,jarra):
        if jarra == 3:
            if self.jarra3l == 0:
                print(">>> No se puede verter la jarra de 3 litros porque esta vacia")
            else:
                while (self.jarra3l != 0 and self.jarra5l < 5):
                    self.jarra5l += 1
                    self.jarra3l -= 1
                print(">>> Verter jarra de 3 litros en la de 5 √")
                self.consultar()

        elif jarra == 5:
            if self.jarra5l == 0:
                print(">>> No se puede verter la jarra de 5 litros porque esta vacia")
            else:
                while (self.jarra5l != 0 and self.jarra3l < 3):
                    self.jarra3l += 1
                    self.jarra5l -= 1
                print(">>> Verter jarra de 5 litros en la de 3 √")
                self.consultar()

        else:
            print(">>> 3 ===> para verter la jarra de 3 litros en la de 5\n>>> 5 ===> para verter la jarra de 5 litros en la de 3")
    
    def consultar(self):
        if self.jarra5l == 4:
            print("                         ¡Lo lograste! Ganaste el juego")
            print("                         ===================================")
            print("                         Jarra de 3 litros ==>",self.jarra3l)
            print("")
            print("                         Jarra de 5 litros ==>",self.jarra5l)
            print("                         ===================================")
            print("")
        else:
            print("")
            print("                         ===================================")
            print("                         Jarra de 3 litros ==>",self.jarra3l)
            print("")
            print("                         Jarra de 5 litros ==>",self.jarra5l)
            print("                         ===================================")
            print("")
```
